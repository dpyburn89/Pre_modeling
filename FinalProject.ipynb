```python

# Here we're going to try a pipeline ### ENDING AT THE KN Neighbors, it works...add models to the pipeline to try and get a best model
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import time
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer

from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from sklearn.neural_network import MLPClassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

y = data['Attrition']
X = data.copy().drop(columns=['Attrition'])


# Initialize lists of string anf numeric variables
str_vars = [] # String variables
num_vars = [] # Numeric variables

# Loop through data.columns and append the column to the apprpriate str_vars or num_vars
# if the conditions are met

for i in X.columns:
    if X.dtypes[i] == 'object':
        str_vars.append(i)
    else:
        num_vars.append(i)



X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

ohe = make_column_transformer(
    (OneHotEncoder(sparse=False), str_vars),
    remainder='passthrough'
)

pipe = Pipeline([("encoder", ohe),
                 ("scaler", StandardScaler()), 
                 ("classifier", LogisticRegression())
                ])

param_grid = [{'classifier': [LogisticRegression()],
               'classifier__C': [.06, .6, 1, 5, 15, 25]},
              {'classifier': [MLPClassifier(activation='relu', solver='adam', max_iter=5000)],
              'classifier__hidden_layer_sizes': [(30, 20), (30, 10),
                                               (60, 40), (60, 20)],
              'classifier__alpha': [0.01, 0.05, 0.1, 0.5, 1]},
              {'classifier': [KNeighborsClassifier()],
               'classifier__n_neighbors': [5,10,15,20,25]},
              {'classifier': [DecisionTreeClassifier()],
               'classifier__max_depth': [1, 3, 7, 10]}]


grid = GridSearchCV(pipe, param_grid=param_grid, cv=5, n_jobs=3)
start_time = time.time()
grid.fit(X_train, y_train)
end_time = time.time()
print("Best cross-validation accuracy: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
print("Test set accuracy: {:.2f}".format(grid.score(X_test, y_test)))
print('Grid search fit took: {0:.02f} s'.format(end_time-start_time))
```

```python
pd.DataFrame(grid.cv_results_).sort_values(by='rank_test_score') # This code provides the results of all the tests run in the grid search above
```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = grid.predict(X_test)
plt.rcParams['figure.figsize'] = (5, 5)
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels=grid.classes_[[0, 1]])
disp.plot(xticks_rotation=90)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
